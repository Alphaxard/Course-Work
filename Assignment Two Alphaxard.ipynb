```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

  # Load the data

```python
data = pd.read_csv("J:/Notes/Datasets/diabetes.csv")  # Read Data
```

  # Calculate the Mean,Skew,Standard deviation , relatioship betwen a varible and the other variables
  # Loop over each column using ehnaced for loop

```python
# Loop over each column using ehnaced for loop
for column in data.columns:
    # a) The center of distribution
    mean = data[column].mean()  # Mean
    mode = data[column].mode()  #Mode
    median = data[column].median() # Median
   
    
    # b) Shape of distribution
    shape = data[column].skew()  # Skewness of the column
    
    # c) Spread of Distribution
    spread = data[column].std()  # Standard deviation of the column
    
    # d) Relationships between any two variables
    correlation = data.corr()[column]  # Correlation with other columns
    
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    sns.histplot(data[column], kde=True)
    plt.title(f"Distribution of {column}")
    
    
    plt.figure(figsize=(8, 6))
    data[column].plot(kind='density')
    plt.title(f'Skewness: {data[column].skew():.2f}')
    plt.xlabel(column)
    plt.show()
    
    plt.show()
      # Displaying  results
    print(f"Column: {column}")
    print("Center of Distribution (Mean):", mean)
    print("Mean", mean)
    print("Mode", mode)
    print("Median", median)
    print("Shape of Distribution (Skewness):", shape)
    print("Spread of Distribution (Standard Deviation):", spread)
    print("Correlations with other numeric columns:")
    print(correlation)
    print("\n")
```
